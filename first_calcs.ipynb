__Predator-Prey Population Dynamics__ <br>
In this program, you enter parameters for the population growth and interactions between a predator and prey population, and then plot the population sizes

```python
# This command invokes the algorithm to solve the differential equation. Do not edit it.

from scipy.integrate import solve_ivp
```

```python
# This cell defines the function that represents predator-prey dynamics. Do not edit.

def lotkavolterra(t,z,a,b,c,d):
    """
    predator-prey dynamics
    """
    x,y = z
    return [a*x - b*x*y, -c*y + d*x*y]
```

```python
# In this cell you can enter values for the parameters of the predator-prey interaction.
a = 2.0   # intrinsic growth rate of prey
b = 1.0   # rate at which predators consume prey
c = 2.0   # rate of predator mortality 
d = 1.0   # rate of conversion of prey to predator offspring
```
